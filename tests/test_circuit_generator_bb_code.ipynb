# 测试circuit_generator代码生成是否合理

d=6,r=6, nkd = [[72,12,6]], q=144。

A_poly=x^3+y+y^2,B_poly=y^3+x+x^2

```python
import stim

circuit_noisy = stim.Circuit.from_file("./bbcode/r=6,d=6,p=0.001,noise=si1000,c=bivariate_bicycle_Z,nkd=[[72,12,6]],q=144,iscolored=True,A_poly=x^3+y+y^2,B_poly=y^3+x+x^2.stim")

circuit_noisy.without_noise().diagram("timeslice-svg", tick = range(0,15))
```

```python
circuit_noisy_1 = circuit_noisy.copy()
for i in range(len(circuit_noisy)):
    if isinstance(circuit_noisy[i], stim.CircuitRepeatBlock):
        circuit_noisy_1.pop(i)
```

```python
circuit_noisy_1.without_noise().diagram("timeslice-svg")
```

```python
detector_error_model = circuit_noisy.detector_error_model()
detector_error_model
```

```python
detector_error_model = circuit_noisy_1.detector_error_model()
detector_error_model
```

```python
circuit_noisy_n = circuit_noisy.copy()
r = 4
n = r-1
for i in range(len(circuit_noisy)):
    if isinstance(circuit_noisy[i], stim.CircuitRepeatBlock):
        REPEAT_part = circuit_noisy_n.pop(i).body_copy()
        REPEAT_part_n = REPEAT_part * n
        circuit_noisy_n.insert(i, REPEAT_part_n)

print(circuit_noisy_n)
```
