```python
import pandas as pd

# 假设数据存储在名为data.csv的文件中，读取数据
qldpc_df_speed = pd.read_csv('/home/normaluser/ck/epmld/experiment/epmld_paper_experiment/result/noisy_varying_qldpc_code_speed_results_n_72_r_d.csv')

# 筛选条件
conditions = (
    qldpc_df_speed['probability'].isin([0.1, 1, 10,100]) &
    (qldpc_df_speed['decoder_method'] == 'EAMLD')
)

# 根据条件筛选数据
qldpc_df_speed = qldpc_df_speed[conditions]

# # 打印结果
# print(result)
```

```python
# 1. 合并X和Z，logical_error_rate取平均值
surface_df_acc = surface_df_acc.groupby(['d', 'round', 'probability', 'noise_model','decoder_method', 'have_stabilizer'],
                               as_index=False)['logical_error_rate'].mean()
qldpc_df_acc = qldpc_df_acc.groupby(['nkd', 'round', 'probability', 'noise_model','decoder_method', 'have_stabilizer'],
                             as_index=False)['logical_error_rate'].mean()

surface_df_speed = surface_df_speed.groupby(['d', 'round', 'probability', 'noise_model','decoder_method', 'have_stabilizer'],
                               as_index=False)['average_per_round_time'].mean()
qldpc_df_speed = qldpc_df_speed.groupby(['nkd', 'round', 'probability', 'noise_model','decoder_method', 'have_stabilizer'],
                             as_index=False)['average_per_round_time'].mean()
```
