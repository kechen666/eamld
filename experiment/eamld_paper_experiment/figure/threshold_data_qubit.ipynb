```python
SCALE_FACTOR = 1.2  # 根据需要调整这个值（1.2-2.0之间）

TITLE_SIZE = 24 * SCALE_FACTOR
LABEL_SIZE = 20 * SCALE_FACTOR
TICK_SIZE = 20 * SCALE_FACTOR
MARKER_SIZE = 20 * SCALE_FACTOR
LINE_WIDTH = 4 * SCALE_FACTOR
LEGEND_SIZE = 20 * SCALE_FACTOR
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.lines import Line2D
from scipy.interpolate import interp1d

# Read the data
df = qldpc_merged_df_acc
# Calculate actual p values
df['p'] = df['probability'] / 10000

# Replace 0.0 with a small value
df['logical_error_rate'] = df['logical_error_rate'].replace(0.0, 1e-10)

# 选择要绘制的方法（这里以EPMLD为例）
selected_method = "EAMLD"  # 可更改为 'MWPM' 或 'Approx_EPMLD' 或者 'EPMLD'

# Create the plot
plt.figure(figsize=(12, 8))
ax = plt.gca()

# 存储各距离的插值函数
interp_funcs = {}

# 只绘制选定方法的数据
for distance in sorted(df['d'].unique()):
    subset = df[(df['d'] == distance) & (df['decoder_method'] == selected_method)]
    if len(subset) > 0:
        # 对数据进行排序
        subset = subset.sort_values('p')
        
        # 创建插值函数（使用对数坐标，因为数据可能跨度很大）
        interp_funcs[distance] = interp1d(subset['p'], np.log10(subset['logical_error_rate']), 
                                        kind='linear', fill_value='extrapolate')
        
        plt.plot(subset['p'], subset['logical_error_rate'],
                label=f'd={distance}',
                color=colors[distance],
                linestyle=linestyles[distance],
                marker=markers[distance],
                markersize=10,
                linewidth=2)  # 加粗线条

# 计算交点函数
def find_intersection(d1, d2):
    # 定义差值函数
    def diff_func(p):
        return interp_funcs[d1](p) - interp_funcs[d2](p)
    
    # 寻找交点
    p_vals = np.linspace(min(df['p']), max(df['p']), 1000)
    diffs = diff_func(p_vals)
    
    # 寻找符号变化的点
    intersections = []
    for i in range(len(p_vals)-1):
        if diffs[i] * diffs[i+1] <= 0:
            # 使用二分法精确查找交点
            from scipy.optimize import bisect
            try:
                root = bisect(diff_func, p_vals[i], p_vals[i+1])
                intersections.append(root)
            except:
                continue
    return intersections

# 计算d=3和d=5的交点
intersections_3_5 = find_intersection(3, 5)
# 计算d=5和d=7的交点
intersections_5_7 = find_intersection(5, 7)

# 绘制交点垂直线
for p in intersections_3_5:
    plt.axvline(x=p, color='gray', linestyle='--', alpha=0.7)
    plt.text(p, plt.ylim()[0] * 1.1, f' p={p:.5f}', rotation=90, 
            verticalalignment='bottom', horizontalalignment='right', fontsize=TITLE_SIZE)

for p in intersections_5_7:
    plt.axvline(x=p, color='gray', linestyle='--', alpha=0.7)
    plt.text(p, plt.ylim()[0] * 1.1, f' p={p:.5f}', rotation=90, 
            verticalalignment='bottom', horizontalalignment='right', fontsize=TITLE_SIZE)

# x_ticks = [0, 0.0001, 0.002,0.004, 0.006, 0.008, 0.01, 0.012, 0.014]
ax.set_xlabel('p', fontsize=LABEL_SIZE)
# ax.set_xticks(x_ticks)
# ax.tick_params(axis='x', which='minor', bottom=False)
ax.tick_params(axis='x', labelsize=TICK_SIZE)

# y_ticks = [0, 0.0001, 0.002,0.004, 0.006, 0.008, 0.01, 0.012, 0.014, 0.016]
ax.set_ylabel('Mean X/Z Logical Error Rate', fontsize=LABEL_SIZE)
# ax.set_yticks(y_ticks)
# ax.set_yticklabels([f"{y:.2f}" if y < 0.1 else f"{y:.1f}" for y in y_ticks], 
                #  fontsize=TICK_SIZE)
ax.tick_params(axis='y', which='major', labelsize=TICK_SIZE)

# plt.xlabel('p', fontsize=32)
# plt.ylabel('Mean X/Z Logical Error Rate', fontsize=32)
# plt.yscale('log')  # 使用对数坐标
plt.grid(True, which="both", ls="--", alpha=0.3)

plt.tight_layout()

save_dir = Path("/home/normaluser/ck/epmld/experiment/epmld_paper_experiment/figure/data_qubits_acc")
save_dir.mkdir(parents=True, exist_ok=True)
save_path = save_dir / "data_threshold.pdf"
plt.savefig(save_path, format="pdf", dpi=300, bbox_inches='tight')

plt.show()
# 打印交点信息
print(f"Intersections between d=3 and d=5: {intersections_3_5}")
print(f"Intersections between d=5 and d=7: {intersections_5_7}")
```
